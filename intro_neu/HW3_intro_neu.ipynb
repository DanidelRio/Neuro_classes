```python
def voltage_p(ra_rm, v_rest=-75, vm=-55):
    return - ra_rm*v_rest + vm*(1 + ra_rm)
```

```python
ra_rm = [i/2+0.5 for i in range(20)]

vp = [voltage_p(ra_rm[i])/(-55) for i in range(len(ra_rm))]
```

```python
plt.plot(ra_rm, vp, color = "black")
plt.xlabel("$r_a/r_m$", fontsize=16)
plt.ylabel("$V_p/V_m$", fontsize=16)
# plt.legend()
plt.show()
```

```python
def relationship(ra_rm, v_rest=-75, vm=-55):
    return 1 + ra_rm*(vm/(vm-vp))
```
